#### DBpedia lookup

```python
import requests
q = "Indian Institute of Technology, Bhilai"
max_results = 5 

response = requests.post(
    f"https://lookup.dbpedia.org/api/search?query={q}&format=json&maxResults={max_results}"
    )
```

##### More query parameters:
- exactMatchBoost
- prefixMatchBoost 
- fuzzyMatchBoost
- fuzzyEditDistance
- fuzzyPrefixLength
- minRelevanceScore 

```python
docs = response.json()['docs']
```

```python
resources = [d['resource'][0] for d in docs]
scores = [d['score'][0] for d in docs]
```

```python
for s,r in zip(scores, resources):
    print(f'Score: {s}, resource: {r}')
```
