```python
import requests
```

```python
q = "Bhilai"
response = requests.post(f"https://lookup.dbpedia.org/api/search?query={q}&format=json&maxResults=")
```

```python
results = response.json()
len(results)
```

```python
print(len(results['docs']))
docs = results['docs']
```

```python
docs
```

```python
resources = [d['resource'] for d in docs]
resources
```

```python
import matplotlib.pyplot as plt
scores = [float(d['score'][0]) for d in docs]
```

```python
plt.plot(scores)
plt.text(0,scores[0],resources[0])
plt.text(15,scores[15], resources[15])
plt.text(99, scores[99], resources[99])
```

```python
plt.hist(scores, bins=10);
```
